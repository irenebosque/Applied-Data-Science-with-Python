```python
Energy = pd.read_excel('Energy Indicators.xls')
Energy = Energy.rename(columns={'Unnamed: 2': 'Country', 'Unnamed: 3': 'Energy Supply', 'Unnamed: 4': 'Energy Supply per Capita', 'Unnamed: 5': '% Renewable'})
cols = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
Energy = Energy[cols] # We don't want the first 2 columns
Energy = Energy.iloc[17:244] # From Afghanistan till Zimbabwe
Energy = Energy.reset_index(drop=True) # Index starts at 0 and not at 17
Energy['Energy Supply'].replace('...', np.nan, inplace=True)
Energy['Country']= Energy['Country'].replace('\d', '', regex=True)

Energy['Country'] = Energy['Country'].replace({"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"})

#Energy['Country'] = Energy['Country'].str.extract('([\w ]+)') #Remove parentheses
Energy['Country'] = Energy['Country'].str.replace(" \(.*\)","",regex=True)
Energy['Energy Supply'] = Energy['Energy Supply']* 1000000
Energy.head()
```

```python
# GPD DataFrame
GDP = pd.read_csv('world_bank.csv', skiprows=4)
GDP['Country Name'] = GDP['Country Name'].replace({"Korea, Rep.": "South Korea", 
"Iran, Islamic Rep.": "Iran",
"Hong Kong SAR, China": "Hong Kong"})
GDP.rename(columns={'Country Name': 'Country'}, inplace=True)
cols = ['Country','2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
GDP = GDP[cols]
GDP.head()
```

```python
# ScimEn
ScimEn = pd.read_excel('scimagojr-3.xlsx', skipfooter=176)
ScimEn.head()
```

```python

#df = pd.merge(ScimEn, Energy, how='inner', left_on='Country', right_on='Country')
df_first_merge = pd.merge(ScimEn,Energy,how='inner',left_on='Country',right_on='Country')

df = pd.merge(df_first_merge, GDP, how="inner",left_on='Country',right_on='Country' )
df = df.set_index(df['Country'])
df = df.drop(columns=['Country'])
print('DataFrame shape: ', df.shape)
df.head()
```
