It would be interesting to see if there is any evidence of a link between vaccine effectiveness and sex of the child. Calculate the ratio of the number of children who contracted chickenpox but were vaccinated against it (at least one varicella dose) versus those who were vaccinated but did not contract chicken pox. Return results by sex.
This function should return a dictionary in the form of (use the correct numbers):
    {"male":0.2,
    "female":0.4}
Note: To aid in verification, the chickenpox_by_sex()['female'] value the autograder is looking for starts with the digits 0.0077.

- HAD_CPOX: kid has had chickenpox. yes = 1, no = 2
- P_NUMVRC – total number of varicella doses
- SEX, male = 1, female = 2

How many girls were vaccinated / How many girls that were vaccinated, had chickpox

```python
import pandas as pd
df = pd.read_csv('NISPUF17.csv')

df = df[df['P_NUMVRC'] >= 1]
yes_CPOX = df[df['HAD_CPOX'] == 1]
no_CPOX  = df[df['HAD_CPOX'] == 2]

male_infected   = len(yes_CPOX[yes_CPOX['SEX'] ==1])
female_infected = len(yes_CPOX[yes_CPOX['SEX'] ==2])

male_no_infected   = len(no_CPOX[no_CPOX['SEX'] ==1])
female_no_infected = len(no_CPOX[no_CPOX['SEX'] ==2])



ratio = {"male":male_infected/male_no_infected,
    "female":female_infected/female_no_infected}
print(ratio)


```
