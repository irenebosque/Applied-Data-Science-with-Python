```python
def result():
    s = 'ACAABAACAAABACDBADDDFSDDDFFSSSASDAFAAACBAAAFASD' #CFB

    result = []
    # compete the pattern below
    pattern = "([^A])(?=AAA)"
    for item in re.finditer(pattern, s):
      # identify the group number below.
      result.append(item.group())
      
    return result
print(result())
```

```python
# Create csv
NHL = pd.read_csv('nhl.csv')
NFL = pd.read_csv('nfl.csv')
MLB = pd.read_csv('mlb.csv')
NBA = pd.read_csv('nba.csv')
```

```python
# Create DataFrame from cities
cities=pd.read_html("wikipedia_data.html")[1]
cities = cities[['Metropolitan area', 'Population (2016 est.)[8]', 'NFL', 'MLB', 'NBA', 'NHL']]
cities.head()
```

```python
# QUESTION 1


# Get only 2018 data
NHL = NHL[NHL['year']== 2018]
```

```python


# Get only columns of interest
NHL = NHL[['team', 'W', 'L']]

# Clean the teams names (remove *)

def remove_star(team):
  if '*' in team:

    return re.findall("([\w\s]*)", team)[0]
  else:
    return team

NHL['team'] = NHL.team.apply(remove_star)

NHL.head()
```

```python
# Clean cities names from the "cities" DatFrame ()
cities

def clean_city_name(city):
  if "[" in city:
    return re.findall('([\w\s]*)', city)[0]
  else:
    return city

cities['NHL'] = cities.NHL.apply(clean_city_name)
cities.head()
```

```python
# In the NHL DataFrame, separate city name from teams name.
# Example: Tampa Bay Lightning -> Tampa Bay, Lightning
NHL['Team only'] = NHL.team.apply(lambda x: x.rsplit(None, 1)[-1])

NHL.loc[3, 'Team only'] = 'Maple Leafs'
NHL.loc[5, 'Team only'] = 'Red Wings'
NHL.loc[13, 'Team only'] = 'Blue Jackets'
NHL.loc[27, 'Team only'] = 'Golden Knights'

NHL.head()

```

```python
# Rename cities column in a more readable format
cities.rename(columns={'Metropolitan area': 'city',  'Population (2016 est.)[8]': 'population','NHL':'Team only'}, inplace=True)
cities = cities[['city', 'population', 'Team only']]
cities.head()
```
